```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Setup Selenium driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

url = "https://internshala.com/internships/data-analytics-internship/"
driver.get(url)
time.sleep(5)  # wait for page to load completely

# Extract job cards
jobs = driver.find_elements(By.CLASS_NAME, "individual_internship")
#job.find_element(By.CLASS_NAME, "heading_4_6")  # or full string "heading_4_6 profile"

job_data = []

for job in jobs:
    try:
        # Example:
        title = job.find_element(By.CSS_SELECTOR, "div.heading_4_6.profile").text
        #title = job.find_element(By.CLASS_NAME, "profile").text
        company = job.find_element(By.CLASS_NAME, "company_name").text
        location = job.find_element(By.CLASS_NAME, "location_link").text
        start_date = job.find_elements(By.CLASS_NAME, "start_date")[0].text
        duration = job.find_elements(By.CLASS_NAME, "duration")[0].text
        stipend = job.find_elements(By.CLASS_NAME, "stipend")[0].text
        skills_tags = job.find_elements(By.CLASS_NAME, "skill_container")
        skills = [skill.text for skill in skills_tags]

        job_data.append({
            "Title": title,
            "Company": company,
            "Location": location,
            "Start Date": start_date,
            "Duration": duration,
            "Stipend": stipend,
            "Skills": skills
        })

    except Exception as e:
        print(f"Error: {e}")

driver.quit()

# Save to DataFrame
df = pd.DataFrame(job_data)
print(df.head())
#df.to_csv("internshala_fresher_jobs.csv", index=False)
```
